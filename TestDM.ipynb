# NumPy Python

Numpy es una biblioteca de Python dedicada al análisis de datos y cálculos numéricos. La biblioteca contiene objetos llamados matrices que ayudan a representar el mismo tipo de recopilación de datos en múltiples dimensiones.

# Plotly en Python

La librería Plotly ayuda a crear gráficos interactivos con calidad. Ejemplos pueden ser diagramas de dispersión, gráficos de áreas, gráficos de barras. 

```python
import plotly.express as px
```

## Area() 

La función area() permite crear un gráfico de área ampliada. La áreas llenas de un valor pertenecen al parámetro line_group.

```python
figura_1 = px.area(df,x="bathrooms",y="price", color="city",line_group="country")
figura_1.show()
```

## Box()

La función box() nos permite ver un diagrama de caja, representa la distribución de la columna ingresada como argumento.

```python
fig = px.box(df, y="price")
fig.show()
```

 ## Strip()

La función strip() ayuda a crear gráficos de bandas utilizando boxtrazos subyacentes.

```python
fig = px.strip(df, x="price", y="date")
fig.show()
```

# Matplotlib en Python

Es una de las librerías más conocidas para la visualización de datos, nos ayuda a producir figuras de calidad, nos brinda una variedad de formatos y entornos interactivos.

```python
import matplotlib.pyplot as plt
```

# Figure ()

La función figure() nos permite crear una nueva figura, y los comando xlabels, ylabels nos permites crear etiquetas en nuestra figura.

```python
plt.figure(figsize = (12, 6))
sns.pairplot(df, x_vars = ['sqft_living'], y_vars = ['price'], height = 7, kind = 'scatter')
plt.xlabel('Pies cuadrados de vivienda')
plt.ylabel('Precio')
plt.title('Pies cuadrados de vivienda vs precio')
plt.show()
```

## Displot()

La función displot() nos brinda un acceso a enfoques para poder visualizar la distribución univariada y de igual forma la bi variada de datos.

```python
sns.displot(
    data=df,
    x="sqft_living", y="price")
```

 ## Relplot()

La función Relplot() brinda diferentes niveles de ejes que muestran la relación entre dos variables.

```python
sns.set_theme()
sns.relplot(
    data=df,
    x="sqft_living", y="price")
```
